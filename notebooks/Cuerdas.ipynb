```python
import pandas as pd

# Especifica la ruta del archivo CSV
ruta_del_archivo = '../data/US.TradeMatrix_771_20240528_163000.csv'

# Carga el archivo CSV en un DataFrame de pandas
df = pd.read_csv(ruta_del_archivo)

# Muestra las primeras filas del DataFrame para verificar que se cargó correctamente
df.head()
```

```python
import pandas as pd
import holoviews as hv
from holoviews import opts, dim
from bokeh.io import show

hv.extension('bokeh')

df['Partner_Label'] = df['Partner_Label'].replace('World', 'Rest of the world')
df['Partner_Label'] = df['Partner_Label'].replace('Russian Federation', 'Rusia')
df['Partner_Label'] = df['Partner_Label'].replace('United States of America', 'USA')
df['Partner_Label'] = df['Partner_Label'].replace('United Kingdom', 'UK')

df['Economy_Label'] = df['Economy_Label'].replace('Russian Federation', 'Rusia')
df['Economy_Label'] = df['Economy_Label'].replace('United States of America', 'USA')
df['Economy_Label'] = df['Economy_Label'].replace('United Kingdom', 'UK')

# Listas de países BRICS y G7
BRICS = ['Brazil', 'Rusia', 'India', 'China', 'South Africa']
G7 = ['Canada', 'France', 'Germany', 'Italy', 'Japan', 'UK', 'USA']

# Función para clasificar los países
def classify_economy(economy):
    if economy in BRICS:
        return 'BRICS'
    elif economy in G7:
        return 'G7'
    else:
        return 'Rest'

# Crear la nueva columna
df['Classification'] = df['Economy_Label'].apply(classify_economy)

# Seleccionar las columnas relevantes
df_95 = df[['Product_Label', 'Economy_Label', 'Partner_Label', '1995_US_dollars_at_current_prices_in_thousands_Value','Classification']]
df_21 = df[['Product_Label', 'Economy_Label', 'Partner_Label', '2021_US_dollars_at_current_prices_in_thousands_Value','Classification']]
```

```python
print(df_95.dtypes)
print(df_21.dtypes)

df_95['1995_US_dollars_at_current_prices_in_thousands_Value'] = df_95['1995_US_dollars_at_current_prices_in_thousands_Value'].fillna(0)
df_95['1995_US_dollars_at_current_prices_in_thousands_Value'] = df_95['1995_US_dollars_at_current_prices_in_thousands_Value'].astype(int)

df_21['2021_US_dollars_at_current_prices_in_thousands_Value'] = df_21['2021_US_dollars_at_current_prices_in_thousands_Value'].fillna(0)
df_21['2021_US_dollars_at_current_prices_in_thousands_Value'] = df_21['2021_US_dollars_at_current_prices_in_thousands_Value'].astype(int)

print(df_95.dtypes)
print(df_21.dtypes)
```

```python
values_to_remove = ['World','Africa', 'Americas', 'Asia', 'Europe', 'Oceania']
df_95_Reduced = df_95[~df['Partner_Label'].isin(values_to_remove)]
df_21_Reduced = df_21[~df['Partner_Label'].isin(values_to_remove)]
```

```python
df_21_USA = df_21_Reduced[df_21_Reduced['Economy_Label'] == "India"]
df_21_USA
```

```python
# Filtrar las filas donde Partner_Label no es "Rest of the world"
df_95_filtered = df_95_Reduced[df_95_Reduced['Partner_Label'] != "Rest of the world"]

# Calcular la suma de los valores para cada Economy_Label, excluyendo "Rest of the world"
sum_by_economy = df_95_filtered.groupby('Economy_Label')['1995_US_dollars_at_current_prices_in_thousands_Value'].transform('sum')

# Agrupar el DataFrame filtrado por Economy_Label y calcular la suma de los valores
sum_by_economy = df_95_filtered.groupby('Economy_Label')['1995_US_dollars_at_current_prices_in_thousands_Value'].sum().reset_index()

# Restar la suma calculada del valor de los paises G7 y BRICS
df_final_95 = df_95_Reduced.merge(sum_by_economy, on='Economy_Label', suffixes=('_original', '_sum_by_economy'))

# Restar los valores cuando Economy_Label del nuevo df sea igual a Economy_Label del df original y Partner_Label sea "Rest of the world"
df_final_95['1995_US_dollars_at_current_prices_in_thousands_Value_original'] = df_final_95.apply(lambda row: row['1995_US_dollars_at_current_prices_in_thousands_Value_original'] - row['1995_US_dollars_at_current_prices_in_thousands_Value_sum_by_economy'] if row['Partner_Label'] == 'Rest of the world' else row['1995_US_dollars_at_current_prices_in_thousands_Value_original'], axis=1)

# Eliminar la columna de suma
df_final_95.drop('1995_US_dollars_at_current_prices_in_thousands_Value_sum_by_economy', axis=1, inplace=True)

# Renombrar la columna
df_final_95.rename(columns={'1995_US_dollars_at_current_prices_in_thousands_Value_original': '1995_US_dollars_at_current_prices_in_thousands_Value'}, inplace=True)

# Mostrar el DataFrame resultante
df_final_95
```
